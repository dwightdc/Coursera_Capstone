### 2. Create a map of Toronto with neighborhoods superimposed on top.

Get Coordinates of centre of Toronto

```python
geolocator = Nominatim(user_agent="Toronto_explorer")
location = geolocator.geocode(query='Toronto, Canada')
TORONTO_LAT_LNG = location.latitude, location.longitude
print('The geograpical coordinates of Toronto are {}, {}.'.format(*TORONTO_LAT_LNG))
```

Create helper function to help with map generation

```python
def gen_map(name, centre_lat_lng, zoom_start, locations_df, marker_func):
    """Generate map and mark locations
    
    NOTE: marker_func takes a row in locations_df and returns a CircleMarker
    """
    target_map = folium.Map(location=centre_lat_lng, zoom_start=zoom_start)
                                      
    # Add red circle as centre of map
    folium.CircleMarker(
        centre_lat_lng,
        radius=12,
        color='red',
        popup=f'Centre of {name}',
        fill=True,
        fill_color='red',
        fill_opacity=0.6
    ).add_to(target_map)

    # add markers to map
    for loc_row in locations_df.itertuples():    
        marker = marker_func(loc_row)
        marker.add_to(target_map)
        
    return target_map
```

Create a map centred around the coordinates of Toronto

```python
# create function to generate markers
def get_marker(nhood_row):
    return folium.CircleMarker(
        [nhood_row.Latitude, nhood_row.Longitude],
        radius=5,
        popup=f'{nhood_row.Neighborhood}, {nhood_row.Borough}',
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False
    )
# create a new map and add markers to the map
toronto_map = gen_map(
    name='Toronto', centre_lat_lng=TORONTO_LAT_LNG, zoom_start=10, 
    locations_df=df_nhoods, marker_func=get_marker
)
toronto_map
```

#### Turn the JSON into a data-frame and filter and process relevant columns

```python
# Flatten json into pandas data frame and extact columns
df_sample_venues = pd.io.json.json_normalize(venues)[['venue.id', 'venue.name', 'venue.categories', 'venue.location.lat', 'venue.location.lng']]

# Extract category name from categories column.
df_sample_venues['venue.categories'] = df_sample_venues['venue.categories'].str[0].str['name']

# Clean column names
df_sample_venues.columns = ['venue_id', 'venue_name', 'category', 'lat', 'lng']

df_sample_venues.head(2)
```

#### Replicate for all the neighbourhoods, i.e., get the list of venues for all neighbourhoods

1. Make a namedtuple, Venue, of venue details to make code more readable

```python
from collections import namedtuple
Venue = namedtuple('Venue', 'id name category lat lng')
```

2. Create a function which accepts a latitude and longitude of a neighborhood
and returns an iterable of Venue rows - each row containing venue_name,
category, lat, and lng

```python
def get_venues(lat, lng, radius, limit):
    url = (
        f'https://api.foursquare.com/v2/venues/explore?client_id={CLIENT_ID}&'
        f'client_secret={CLIENT_SECRET}&ll={lat},{lng}&'
        f'v={VERSION}&radius={radius}&limit={LIMIT}'
    )
    results = requests.get(url).json()
    for venue in  results['response']['groups'][0]['items']:
        yield Venue(
            id=venue['venue']['id'], 
            name=venue['venue']['name'], 
            category=venue['venue']['categories'][0]['name'],
            lat=venue['venue']['location']['lat'], 
            lng=venue['venue']['location']['lng']
        )  
```

3. Create data-frame of neighbourhoods and venues using df_nhoods and the get_venues() function

```python
rows = (
    (h.Neighborhood, h.Latitude, h.Longitude, v.id, v.name, v.lat, v.lng, v.category)
    for h in df_nhoods.itertuples()
    for v in get_venues(h.Latitude, h.Longitude, radius=1000, limit=100)
)
columns = ['Neighborhood', 'Neighborhood Lat', 'Neighborhood Lng',
           'Venue Id', 'Venue', 'Venue Lat', 'Venue Lng', 'Venue Category']
df_nhood_venues = pd.DataFrame(data=rows, columns=columns)
print(df_nhood_venues.shape)
df_nhood_venues.head()
```

#### Create a data-frame of neighbourhoods and one-hot encoded columns of venue categories in that neighborhood

1. Rename Neighborhood to Neighborhood_Name

```python
df_nhood_venues.rename(columns={'Neighborhood': 'Neighborhood_Name'}, inplace=True)
```

2. Perform one-hot encoding

```python
to_concat = [df_nhood_venues['Neighborhood_Name'], pd.get_dummies(df_nhood_venues['Venue Category'])]
df_nh_venues_onehot = pd.concat(to_concat, axis=1)
print(df_nh_venues_onehot.shape)
df_nh_venues_onehot.head()
```

#### Find mean of each column, grouped by neighborhood. This gives the percentage of that venue category in the neighborhood

```python
df_nh_venues_grouped = df_nh_venues_onehot.groupby('Neighborhood_Name').mean().reset_index()
print(df_nh_venues_grouped.shape)
df_nh_venues_grouped.head()
```

#### Create a data frame of most common venues sorted by percentage of venues in neighbourhood

```python
# Helper functions
def top_n_columns(limit):
    """Fucntion to create 'TOP-N' column names"""
    ords = {1:'st', 2:'nd', 3:'rd'}
    return [
        f'{i}{ords.get(i, "th")} Most Common Venue'
        for i in range(1, limit+1)
    ]

def top_n_venues(venue_freq_ser, limit=10):
    """Sort percentages of frequences and return corresponding venue categories"""
    return venue_freq_ser.sort_values(ascending=False).head(limit).index.values

# Create data-frame of top 10 venues category names per neighbourhood
top_n_limit = 10
df_nh_top10_venues_sorted = df_nh_venues_grouped.drop('Neighborhood_Name', axis=1)
df_nh_top10_venues_sorted = df_nh_top10_venues_sorted.apply(top_n_venues, axis=1, result_type='expand', limit=top_n_limit)
df_nh_top10_venues_sorted.columns = top_n_columns(limit=top_n_limit)
df_nh_top10_venues_sorted.insert(0, value=df_nh_venues_grouped['Neighborhood_Name'], column='Neighborhood_Name')
print(df_nh_top10_venues_sorted.shape)
df_nh_top10_venues_sorted.head()
```

### Cluster neighbourhoods

#### Use elbow method to find the optimal number of clusters visually

```python
# Create a data frame with only the percentages
df_nh_venues_clustering = df_nh_venues_grouped.drop('Neighborhood_Name', axis=1)

min_k, max_k, = 2, 20
ks = range(min_k, max_k)
kmeans_list = [
    KMeans(n_clusters=k, random_state=0).fit(X=df_nh_venues_clustering)
    for k in ks    
]
sum_of_squared_distances = [kmeans.inertia_ for kmeans in kmeans_list]
plt.figure(figsize=(15,7))
plt.plot(ks, sum_of_squared_distances, 'bx-')
plt.xlabel('k')
plt.ylabel('sum_of_squared_distances')
plt.xticks(ks)
plt.title('Elbow Method For Optimal k')
```

#### There is a slight elbow at k=6, therefore, we get the labels of the kmeans object with a k of 6, and add the labels to the sorted venues data frame

```python
N_CLUSTERS = 6
labels = kmeans_list[N_CLUSTERS-min_k].labels_
print('Sorted Lables:\n', np.sort(labels))
df_nh_venue_clusters = pd.concat([pd.Series(labels, name='Cluster_Labels'), df_nh_top10_venues_sorted], axis=1)
df_nh_venue_clusters.head()
```

#### Merge with neighbourhood data to get latitudes and longitudes

```python
df_nh_merged = df_nhoods.merge(df_nh_venue_clusters, left_on='Neighborhood', right_on='Neighborhood_Name').drop('Neighborhood', axis=1)
```

#### Visualize Clusters

```python
# Create N_CLUSTERS number of rainbow colors
colors_array = cm.rainbow(np.linspace(0, 1, N_CLUSTERS))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# create function to generate markers
def get_marker(nhood_row):
    return folium.CircleMarker(
        [nhood_row.Latitude, nhood_row.Longitude],
        radius=5,
        popup=f'{nhood_row.Neighborhood_Name}, Cluster {nhood_row.Cluster_Labels}',
        color=rainbow[nhood_row.Cluster_Labels],
        fill=True,
        fill_color=rainbow[nhood_row.Cluster_Labels],
        fill_opacity=0.7,
        parse_html=False
    )
# create a new map and add markers to the map
toronto_map_clusters = gen_map(
    name='Toronto', centre_lat_lng=TORONTO_LAT_LNG, zoom_start=10, 
    locations_df=df_nh_merged, marker_func=get_marker
)
toronto_map_clusters
```
